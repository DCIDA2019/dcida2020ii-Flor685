$\textbf{*** Find the cumulative distribution for the Gaussian Distribution, and reproduce the plots.}$

La CMD está dada para distribución Gaussiana por: $$ F_g(x)=\int_{-\infty}^{x} \frac{e^{-x'^{2}/2}}{2\pi} dx' $$
No puede ser escrita en términos de funciones analíticas.

```python
#Para graficar la función
import matplotlib.pyplot as plt
import numpy as np
from scipy import integrate

import math

yy=np.linspace(-10,10,300)


def fg(x):
    return math.exp(-x**2 *0.5)/2*math.pi

ploth=[(integrate.quad(fg,math.inf,t)[0]) for t in yy]
plt.plot(yy,ploth,color='green')
plt.show()
```

$\textbf{*** For the PDFs mentioned, find CDF, mean and standar deviation. Ploth both the PDF and CDF, for some diferents values of mean and sigma.}$

Para la distribución de $\textbf{POISSON}$, la PDF, con k es el número de ocurrencias del evento o fenómeno y λ  el número de veces que se espera que ocurra el fenómeno durante un intervalo dado: $$ f(k,\lambda)= \frac{e^{-\lambda}\lambda^k}{k!} $$.
La CDF está dada por:
$$ \frac{\Gamma([k+1], \lambda)}{k!} $$ para $k>0$

```python
from scipy import stats

l=4
poisson = stats.poisson(l) # Distribución
x = np.arange(poisson.ppf(0.01),
              poisson.ppf(0.99))
fmp = poisson.pmf(x) # Función de Masa de Probabilidad
plt.plot(x, fmp, '--')
plt.vlines(x, 0, fmp, colors='b', lw=5, alpha=0.5)
plt.title('Distribución Poisson')
plt.ylabel('probabilidad')
plt.xlabel('valores')
plt.show()


x_1 = np.linspace(stats.norm(10, 1.2).ppf(0.01),
                  stats.norm(10, 1.2).ppf(0.99), 100)
fda_poisson = stats.poisson.cdf(x,l) # Función de Distribución Acumulada
fda_poisson2 = stats.poisson.cdf(x,5)
fda_poisson3 = stats.poisson.cdf(x,1)
plt.plot(x, fda_poisson, '--', label='CDF poisson')
plt.plot(x, fda_poisson2, '-')
plt.plot(x, fda_poisson3, 'o')
plt.title('Función de Distribución Acumulada')
plt.ylabel('probabilidad')
plt.xlabel('valores')
plt.legend(loc=4)
plt.show()
```

En la distribución $\textbf{BETA}$, PDF, donde $B(\alpha,\beta)$ es la función Beta :
    $$ \frac{x^{\alpha^-1}(1-x)^{\beta -1}}{B(\alpha,\beta)}$$


```python
a, b = 2, 5
beta = stats.beta(a, b)
x = np.linspace(beta.ppf(0.01),
                beta.ppf(0.99), 100)
fp = beta.pdf(x) # Función de Probabilidad
plt.plot(x, fp)
plt.title('Distribución Beta')
plt.ylabel('probabilidad')
plt.xlabel('valores')
plt.show()

x_1 = np.linspace(stats.norm(10, 1.2).ppf(0.01),
                  stats.norm(10, 1.2).ppf(0.99), 100)
fda_beta = stats.beta.cdf(x, a, b) # Función de Distribución Acumulada
fda_beta2 = stats.beta.cdf(x, 3.4, 1)
fda_beta3 = stats.beta.cdf(x, 0.5, 0.5)
plt.plot(x, fda_beta, '--', label='CDF beta')
plt.plot(x, fda_beta2, '-')
plt.plot(x, fda_beta3, '.')
plt.title('Función de Distribución Acumulada')
plt.ylabel('probabilidad')
plt.xlabel('valores')
plt.legend(loc=4)
plt.show()
```

$\textbf{*** If $$ y=\Phi (x) = exp(x) $$ and $$ p(x)=1 $$ para $$ 0\leq x \leq 1$$. What is the resultant distribution for y? }$

La forma para transformar variables es:
$$ p(y)=p[\Phi ^{-1}(y)] \left| \frac{d\Phi ^{-1} (y)}{dy} \right|  $$
Identificando cada parte de la ecuación para nuestro caso: $ \Phi ^{-1}(y)= \log{x} $, entonces $\frac{d\Phi ^{-1} (y)}{dy}=\frac{1}{y}$; y como $p(x)=1$, $p[\Phi ^{-1}(y)]=1 $. Por lo tanto:
$$ p(y)=\left| \frac{1}{y} \right| $$

$\textbf{*** Show explicitly that the linear least square method is derived from the minimization of the chi-squared when the model is a straight line.}$

La función $\chi^2$ es de la forma:
$$ \chi^2= \frac{\sum (y_i -y(x_i ,\theta))^2}{\sigma_{yi} ^2} $$
Para el ajuste lineal $y(x_i ,\theta)= bx+a$, donde b y a son los parámetros. Sustituyendo explícitamente en la ecuación para $chi^2$
$$ \chi^2= \frac{\sum (y_i -bx-a)^2}{\sigma_{yi} ^2} $$
Tenemos que minimizar la función anterior, es decir, encontrar $\frac{\partial \chi ^2}{\partial a }=0$ y $\frac{\partial \chi ^2}{\partial b }=0$. Con lo que obtenemos, correspondientemente:
$$\frac{-2\sum (y_i -bx_i -a)}{\sigma ^2}=0 $$ $$ \frac{-2\sum (y_i -bx_i -a)x_i}{\sigma ^2}=0 $$
Reescribiendo:
$$ \sum (a+bx_i) =\sum y_i $$ $$ \sum (a+bx_i)x_i=\sum x_i y_i $$
Multiplicando la segunda ecuación por $-\sum 1= -n$ y la primera por $\sum x_i$, restando ambas y despejando, obtenemos la ecuación para b:
$$ b= \frac{\sum y_i - n\sum x_i y_i }{(\sum x_i )^2-n\sum x_i ^2} $$
Similarmente podemos obtener a:
$$ a= \frac{\sum x_i y_i \sum x_i - \sum x_i ^2 \sum y_i }{(\sum x_i )^2-n\sum x_i ^2}$$
Usando la expresión para la media obtenemos el resultado presentado en clase.

$\textbf{***}$ Write down a poisson likelihood, and identify wich parameter(s) correspond to the data and wich to the model.

La distribución de poisson:
    $$ f(x, \lambda)= \frac{\lambda^x e^{-\lambda}}{x!} $$
   
La función likelihood
$$ \mathcal{L}(\lambda;x_1,...,x_n) = \prod _{j=1} ^n f(\lambda,x_j) $$
Así que para nuestra función de probablilidad:
$$ \mathcal{L}(\lambda;x_1,...,x_n) = \prod _{j=1} ^n exp(-\lambda)\frac{1}{x_j !} \lambda ^{x_j} $$

El parámetro a encontrar es $\lambda$ que maximiza esta función.
